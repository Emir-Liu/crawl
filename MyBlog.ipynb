# 导入urllib模块

```python
import urllib.request
from bs4 import BeautifulSoup
import re
```

抓取URL内容，也就是发送一个GET请求到指定的页面，然后返回HTTP的相应,然后打印网页中的原始内容。

```python
url = "https://emir-liu.github.io"
page = urllib.request.urlopen(url)
html = page.read()
#print(html)
```

使用bs4结构化信息

```python
obj = BeautifulSoup(html)
print(obj.prettify())
```

将所有的链接处理为set格式的数据，将上面所有的连接标签进行处理为http、www和/输出

```python
def findInternalLinks(obj,links):
    for link in obj.find_all('a'):
        #print(link)
        if link.attrs is not None:
            #print(link.attrs)
            if link.has_attr('href') :
                #print(link['href'])
                match_key = "^(http|www|/)"
                if link['href'] not in links and re.match(match_key,link['href']):
                    links.append(link['href'])
    return 0
links = []
findInternalLinks(obj,links)
print(links)
```

遍历网址set，如果遍历次数小于等于set的大小，然后就检查网址是否为内部网址，如果是则检查其中的内容，然后将新的网址加入set中，然后继续遍历。

寻找所有外部的网址，以www或者http为开头

```python
import urllib.request
from urllib.parse import quote
from bs4 import BeautifulSoup
import re
import string

def findInternalLinks(obj,links):
    for link in obj.find_all('a'):
        #print(link)
        if link.attrs is not None:
            #print(link.attrs)
            if link.has_attr('href') :
                #print(link['href'])
                match_key = "^(http|www|/)"
                if link['href'] not in links and re.match(match_key,link['href']):
                    links.append(link['href'])
    return 0

url = "https://emir-liu.github.io"
links = []
links.append(url)
num_link = 0
while (num_link <= len(links)):
    print('start...')
    #print('num_link:',num_link)
    #print("len_links",len(links))
    new_url = links[num_link]
    num_link = num_link + 1
    internal_key = "^(/)"
    external_key = "^(http|www)"
        
    if re.match(internal_key,new_url) or re.match(url,new_url):
        if not re.match(url,new_url):
            act_url = url+new_url
        else:
            act_url = new_url
        #下面的部分代码还有问题，中文url无法打开
        '''
        print('act_url',act_url)
        urllib.parse.quote(act_url,safe=string.printable)
        print('quote_act_url',act_url)
        '''
        #当url中有中文的时候需要使用上面的函数
        try:
            page = urllib.request.urlopen(act_url)
        except:
            print("could not open %s",act_url)
            continue
        html = page.read()
        obj = BeautifulSoup(html)
        findInternalLinks(obj,links)

    #print('end...')
    #print('links',links)
    print('num_link:',num_link)
    print("len_links",len(links))
print(links)
```
